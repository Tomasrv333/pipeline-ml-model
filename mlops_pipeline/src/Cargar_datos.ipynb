```python
import os
import pandas as pd

# Logs iniciales
print("=== [1] Iniciando carga de datos ===")

# Verificar entorno virtual activo
venv = os.getenv('VIRTUAL_ENV')
if venv:
    print(f"Entorno virtual detectado: {venv}")
else:
    print("⚠️ No se detectó entorno virtual. Asegúrate de ejecutar este notebook en el kernel del entorno creado por set_up.bat.")
```

```python
# =========================
# 2️⃣ Configuración de rutas
# =========================
# Buscar archivo CSV
base_path = os.path.abspath(os.path.join(os.getcwd(), "..", ".."))
csv_path = os.path.join(base_path, "Base_de_datos.csv")

print(f"Ruta esperada del dataset: {csv_path}")

if not os.path.exists(csv_path):
    raise FileNotFoundError(f"❌ No se encontró el archivo Base_de_datos.csv en la ruta: {csv_path}")
else:
    print("✅ Archivo CSV encontrado correctamente.")
```

```python
# =========================
# 3️⃣ Cargar los datos
# =========================
print("=== [2] Cargando archivo CSV... ===")

try:
    df = pd.read_csv(csv_path, encoding="utf-8")
    print("✅ Archivo cargado exitosamente con codificación UTF-8.")
except UnicodeDecodeError:
    df = pd.read_csv(csv_path, encoding="latin-1")
    print("⚠️ Archivo cargado con codificación LATIN-1 (errores de codificación en UTF-8).")
```

```python
# =========================
# 4️⃣ Validación básica
# =========================
print("=== [3] Validando estructura del dataset... ===")
print(f"Dimensiones del dataset: {df.shape[0]} filas, {df.shape[1]} columnas.")
print("\nColumnas detectadas:")
print(df.columns.tolist())

# Mostrar primeras filas
print("\nVista previa de los datos:")
display(df.head())
```

```python
# =========================
# 5️⃣ Validación de valores nulos
# =========================
print("=== [4] Validando valores nulos... ===")
nulls = df.isnull().sum()
print(nulls[nulls > 0] if nulls.sum() > 0 else "✅ No se encontraron valores nulos.")
```
